## Form Generate

Form ini sudah disederhanakan. Hanya ada 3 pilihan:
- **Category**: Topik konten yang diinginkan
- **Images**: Jumlah gambar yang mau di-generate (1-10)
- **Size**: Ukuran gambar (1920×1080 landscape, 2000×2000 square, 1080×1920 portrait)

Parameter lainnya sudah otomatis optimal:
- Steps: 50 (untuk kualitas tinggi)
- Guidance: 7.5 (untuk prompt adherence)
- Upscale: minimal 2000px di sisi terkecil


```python
from agent import PipelineRuntimeOptions, display_runtime_widgets

runtime_widgets = display_runtime_widgets(PipelineRuntimeOptions())
```

```python
from agent import AgentPipeline, runtime_from_widgets
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')

runtime = runtime_from_widgets(runtime_widgets)
pipeline = AgentPipeline.from_runtime_options(runtime)
result = pipeline.run()

print('\n' + '='*60)
print('GENERATION COMPLETED')
print('='*60)
print(f'\nTrend Category: {result.trend_result.category}')
print(f'Trend Source: {result.trend_result.source}')
print(f'Trend Keywords: {result.trend_result.keywords[:5]}...')
print(f'\nPrompt Title: {result.prompt_bundle.title}')
print(f'Prompt Style: {result.prompt_bundle.style}')
print(f'\nSession Folder: {result.drive_paths.session_output}')
print(f'Images Folder: {result.drive_paths.session_images}')
print(f'Metadata File: {result.metadata_csv_path}')
print(f'\nGenerated {len(result.generated_files)} image(s):')
for i, file_path in enumerate(result.generated_files, 1):
    print(f'  {i}. {file_path.name}')
print('\n' + '='*60)
```
